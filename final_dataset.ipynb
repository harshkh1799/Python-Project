```python
from faker import Faker
import pandas as pd
from datetime import datetime, timedelta 
import random
```

```python
fake = Faker('en_US') 
```

```python
# initalizing the dates to for data of 12 months
start_date = datetime(2022,1,1)
end_date = datetime(2022,12,31)
```

```python
# class to create the final dataframe for the count of each leads after the funnel

class ListContainer:
    def __init__(self):
        self.lead_src = []
        self.count_of_leads = []

    def append_to_list(self, cnt, src):
        self.lead_src.append(src)
        self.count_of_leads.append(cnt)

    def return_list(self):
        return self.lead_src, self.count_of_leads


obj = ListContainer()
```

```python
# lead_status_data = ["Warm", "Cold", "MQL", "SQL's", "Hot", "Cusotmer"]
lead_source_data = ["Inbound", "Outbound", "Referals", "Event/Conference", "Webinar/Workshop", "Social Media", "Paid Ad", "Content Marketing",
                     "Email Marketing"]
regions = ['North', 'South', 'East', 'West']

lead_id = []
lead_source = []
date_created = []
name = []
region = []
Cost_Per_Lead = []
lead_status = []


for i in range(1,7001):
    
    names = fake.name()
    date_created_ = fake.date_between(start_date=start_date, end_date=end_date)

#selecting random lead_source - regions
    rand_lead_source = random.choice(lead_source_data)
    rand_cost_per_lead = random.uniform(10, 50)
    rand_region = random.choice(regions)
    
# appending data to respective list
    lead_id.append(i)
    lead_source.append(rand_lead_source)
    date_created.append(date_created_)
    name.append(names)
    region.append(rand_region)
    lead_status.append("Cold")

final_data_1 = {"Lead_id": lead_id,
              "Lead_source": lead_source,
              "Date_Created": date_created,
              "Name": name,
              "Region": region,
              "Cost_Per_Lead": rand_cost_per_lead,
              "lead_status": lead_status}
```

```python
# lead_status_data = ["Warm", "Cold", "MQL", "SQL's", "Hot", "Customer"]
start_date = datetime(2022,1,1)
end_date = datetime(2022,12,31)
num_records = 100

# initalizing the list 
lead_source_data = ["Inbound", "Outbound", "Referals", "Event/Conference", "Webinar/Workshop", "Social Media", "Paid Ad", 
                    "Content Marketing","Email Marketing"]
regions = ['North', 'South', 'East', 'West']
designations = ["Business","Student","Self-Employed","Freelancer","Entrepreneur","Homemaker","Retiree","Working",
                "Consultant","Job_Seeker"]
social_elements=['Twitter', 'Instagram', 'LinkedIn', 'Facebook']

# creating list of defined records for intial stage using list comphresion 
lead_id = [i for i in range(1,num_records+1)]
lead_source = [random.choice(lead_source_data) for _ in range(num_records)]
date_created = [fake.date_between(start_date=start_date, end_date=end_date) for _ in range(num_records)]
name = [fake.name() for _ in range(num_records)]
phone = [random.choice(['9','8','6']) + fake.numerify('#########')  for _ in range(num_records)]
designation = [random.choice(designations) for _ in range(num_records)]
social_platform = [fake.random_element(social_elements) for _ in range(num_records)]
region = [random.choice(regions) for _ in range(num_records)]
Cost_Per_Lead = [random.uniform(10,50) for _ in range(num_records)]
lead_status = ["Cold" for _ in range(num_records)] 


final_data = {"Lead_id": lead_id,
              "Lead_source": lead_source,
              "Date_Created": date_created,
              "Name": name,
              "Phone": phone,
              "Designation": designation,
              "Social_platform": social_platform,
              "Region": region,
              "Cost_Per_Lead": Cost_Per_Lead,
              "lead_status": lead_status}
```
